```python
import whisper
import os

model = whisper.load_model("base")

# Define a list of audio file paths
audio_files = [
    "/content/drive/MyDrive/Recording (14).m4a",
    "/content/drive/MyDrive/Recording (10).m4a",
    "/content/drive/MyDrive/Recording (11).m4a",
    "/content/drive/MyDrive/Recording (12).m4a",
    "/content/drive/MyDrive/Recording (13).m4a",
    "/content/drive/MyDrive/Recording (15).m4a",
    "/content/drive/MyDrive/Recording (16).m4a",
    "/content/drive/MyDrive/Recording (17).m4a",
    "/content/drive/MyDrive/Recording (18).m4a",
    "/content/drive/MyDrive/Recording (19).m4a",

]

for audio_file in audio_files:
    # Load audio file
    audio = whisper.load_audio(audio_file)
    audio = whisper.pad_or_trim(audio)

    # Extract features
    mel = whisper.log_mel_spectrogram(audio).to(model.device)

    # Detect language
    _, probs = model.detect_language(mel)
    detected_language = max(probs, key=probs.get)
    print(f"Detected language for {os.path.basename(audio_file)}: {detected_language}")

    # Decode and print result
    options = whisper.DecodingOptions(fp16=False)
    result = whisper.decode(model, mel, options)
    print(f"Result for {os.path.basename(audio_file)}: {result.text}")
    print("=" * 50)

    translated_result = translator.translate(result.text, src='en', dest='te')
    translated_text = translated_result.text
    print("Translated text:", translated_text)

    # Save translated text as an audio file
    tts = gTTS(translated_text, lang='te')
    output_audio_file = f"translated_{os.path.basename(audio_file)}.mp3"
    tts.save(output_audio_file)

    # Display the translated audio
    ipd.display(ipd.Audio(output_audio_file))
    print("=" * 50)
```

```python
import whisper
import os
from googletrans import Translator
from gtts import gTTS
import IPython.display as ipd

model = whisper.load_model("base")

# Define a list of target languages
target_languages = ['te', 'fr', 'hi', 'de']  # Add more languages as needed

# Define a list of audio file paths
audio_files = [
    "/content/drive/MyDrive/Recording (14).m4a",
    "/content/drive/MyDrive/Recording (10).m4a",
    "/content/drive/MyDrive/Recording (11).m4a",
    "/content/drive/MyDrive/Recording (12).m4a",
    "/content/drive/MyDrive/Recording (13).m4a",
    "/content/drive/MyDrive/Recording (15).m4a",
    "/content/drive/MyDrive/Recording (16).m4a",
    "/content/drive/MyDrive/Recording (17).m4a",
    "/content/drive/MyDrive/Recording (18).m4a",
    "/content/drive/MyDrive/Recording (19).m4a",
]

for audio_file in audio_files:
    # Load audio file
    audio = whisper.load_audio(audio_file)
    audio = whisper.pad_or_trim(audio)

    # Extract features
    mel = whisper.log_mel_spectrogram(audio).to(model.device)

    # Detect language
    _, probs = model.detect_language(mel)
    detected_language = max(probs, key=probs.get)
    print(f"Detected language for {os.path.basename(audio_file)}: {detected_language}")

    # Decode and print result
    options = whisper.DecodingOptions(fp16=False)
    result = whisper.decode(model, mel, options)
    print(f"Result for {os.path.basename(audio_file)}: {result.text}")
    print("=" * 50)

    # Translate and save audio for each target language
    for target_lang in target_languages:
        translator = Translator()
        translated_result = translator.translate(result.text, src='en', dest=target_lang)
        translated_text = translated_result.text

        # Save translated text as an audio file
        tts = gTTS(translated_text, lang=target_lang)
        output_audio_file = f"translated_{os.path.basename(audio_file)}_{target_lang}.mp3"
        tts.save(output_audio_file)

        # Display the translated audio
        ipd.display(ipd.Audio(output_audio_file))
        print(f"Translated text to {target_lang}: {translated_text}")
        print("=" * 50)
```
